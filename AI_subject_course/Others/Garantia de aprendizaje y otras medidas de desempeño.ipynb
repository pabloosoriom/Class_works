```python
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d 
from sklearn import datasets
import random 
import pandas as pd
import seaborn as sns
import plotly_express as px
import numpy as np
import math as m
```

```python
#Probably approximated correct learning guarantee
def pac_learning_guarantee (epsilon,H,delta):
    return (1/epsilon)*(np.log(H)+np.log(1/delta))
```

```python
eps_id=0.9
delta_id=0.8
vec_h=range(1,30)

hs=[]
for i in vec_h:
    hs.append(pac_learning_guarantee(eps_id,i,delta_id))
    
    
plt.plot(vec_h, hs, color='red')
plt.title('Variation of H with Epsilon and Delta cons')
plt.show()
```

```python
eps_id=np.linspace(0.001, 1, 100)
delta_id=0.8
vec_h=5
hs=[]

for i in eps_id:
    hs.append(pac_learning_guarantee(i,vec_h,delta_id))
    
    
plt.plot(eps_id, hs, color='red')
plt.title('Variation of Epsilon with H and Delta cons')
plt.show()
```
